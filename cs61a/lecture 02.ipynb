```python
def fib(n):
    curr= 0
    next = 1
    i = 0
    
    while i<n: 
        i=i+1
        next_next=curr+next
        curr= next
        next= next_next
    return curr
```

```python
fib(1)
```

```python
fib(2)
```

```python
fib(3)
```

```python
def both_positive(x, y):
    """Returns True if both x and y are positive.

    >>> both_positive(-1, 1)
    False
    >>> both_positive(1, 1)
    True
    """
    if x > 0 and y > 0:
        return True
    else:
        return False
```

```python
both_positive(-1, 1)
```

```python
both_positive(1, 1)
```

```python
def sum_digits(n):
    """Sum all the digits of n.

    >>> sum_digits(10) # 1 + 0 = 1
    1
    >>> sum_digits(4224) # 4 + 2 + 2 + 4 = 12
    12
    >>> sum_digits(1234567890)
    45
    >>> x = sum_digits(123) # make sure that you are using return rather than print
    >>> x
    """
    number=0
    total=0
    while number >=0:
        number= n%10
        total = number+ total
        next_number = n //10
        n=next_number
        
    return total  
        
```
